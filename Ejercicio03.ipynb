```haskell
cuadrado :: Int -> Int
cuadrado x = x * x

suma :: Int -> Int -> Int
suma x y = x + y

sumaCuadrados :: Int -> Int -> Int
sumaCuadrados x y = suma (cuadrado x) (cuadrado y)
```

```haskell
duplicar :: Int -> Int
duplicar x = x * 2

dobleLista :: [Int] -> [Int]
dobleLista = map (*2) 
```

```haskell
clasificarNota :: Float -> String
clasificarNota nota 
  | nota < 3 = "Reprobado"
  
  | nota >= 3 && nota < 4 = "En recuperación"
  
  | nota >= 4 = "Aprobado"
  
```

```haskell
fibonacci :: Int -> Int
fibonacci 0 = 0
fibonacci 1 = 1
fibonacci n = fibonacci (n-1) + fibonacci (n-2)
```

```haskell
sumaCuadrados 3 4                  
dobleLista [1,2,3,4,5]     
clasificarNota 2                    
fibonacci 0                
fibonacci 7                
```
