**¿Qué significa que Haskell sea un lenguaje fuertemente tipado y evaluado perezosamente?** Que cada expresión está bien definida en un tiempo específico de ejecución, lo que hace que prevenga los errores de tipo. Las variables no se evalúan hasta qeu sean estrictamente necesarias.

```haskell
doble :: Int -> Int
doble n = n * 2

esMayorEdad :: Int -> Bool
esMayorEdad edad = edad >= 18

areaCirculo :: Float -> Float
areaCirculo radio = pi * radio^2
```

**¿Qué diferencias encuentras entre los tipos Int y Float?**
Int representa todos los valores enteros con un rango fijo, mientras que float puede representar números decimales.

```haskell
-- Pruebas
doble 8
esMayorEdad 17
areaCirculo 2.0
```
